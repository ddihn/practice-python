### 행 단위 데이터 수집 실습

```python
import requests, bs4
import pandas as pd

# 웹페이지 전체 html 소스 가져오기
resp = requests.get("https://sparkkorea.com/%ed%80%b4%ec%a6%88/")
resp.encoding = "utf-8"
html = resp.text
bs = bs4.BeautifulSoup(html,'html.parser')

# 퀴즈 부분 Div 태그 추출
quizDiv = bs.find(name="div",attrs={"id":"id_spark_quiz"})

# a 태그 추출
quizInAtags = quizDiv.find_all(name="a")

# 행으로 저장할 list 선언
quizList = []

# 속성들을 행으로 저장
for i in range(len(quizInAtags)) :
    eachTags = quizInAtags[i]
    eachTagText = eachTags.text
    eachTagLink = eachTags.attrs["href"]
    quizList.append([eachTagText,eachTagLink])

# 행들이 저장된 list를 DataFrame형태로 변환
quizDf = pd.DataFrame(quizList,columns =["Text","Link"])

quizDf
```
