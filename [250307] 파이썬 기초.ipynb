```python
# 사칙연산
intValue = 10
floatValue = 10.5
sumValue = intValue + floatValue # 더하기 연산
divValue = intValue / floatValue # 나누기 연산
multiValue = intValue * floatValue # 곱하기 연산
modValue = intValue % floatValue # 나머지 연산 => 패턴을 잡아내기 용이
expValue = intValue**3 # 제곱 연산 => 데이터를 증폭시키는 데 사용
print(sumValue, divValue, multiValue,modValue)
```

####문자

```python
# 문자열에서 +는 연결의 의미
stringValue = "smartFinance"
type(stringValue)

headChar = "smart"
tailChar = "Finance"
total = headChar+tailChar
print(total)
```

```python
# indexing : 내가 원하는 값만 추출
# 음,양 둘 다 사용가능하나 웬만하면 양의 인덱스 사용 추천
currYearweek = "202509"
yearTerm = 4 # 하드코딩을 피하기 위한 변수 설정
currYearweek[0] #'2'
currYearweek[0:4] #'2025', 0 생략 가능
currYearweek[:4] # 처음부터 원하는 범위까지
currYearweek[4:] # 원하는 범위부터 마지막까지
currYearweek[yearTerm:] # 의미있는 변수라면 변수로 설정하는 것이 유지보수에 좋음
```

```python
# 검색 문자의 출현 횟수 찾기 : count 함수
# 구분자 확인할 때 용이
targetLetters = "LED_LED_TV"
searchWord = "LED_"
targetLetters.count(searchWord)
```

```python
# 원하는 문자의 시작 인덱스 가져오기
inYearweek = "2023W01"
delimeter = "W" # 만약 소문자라면? -> 예외처리 필요
inYearweek = inYearweek.upper() # 전체 대문자로 변경
# 데이터 전처리, 가공 과정은 되도록 코드 앞 단에 위치 시키는 것이 좋음

delimeterIndex = inYearweek.index(delimeter)
inYear = inYearweek[:delimeterIndex]
inWeek = inYearweek[delimeterIndex+1:]
print(f"현재 년도는 {inYear}년 입니다.") # f(formatting)를 붙이면 {변수값} 출력 가능
print(f"현재 주차는 {inWeek}주차 입니다.")
```

```python
# 모든 구분자를 "_"로 변경 : replace()
# 공백을 제거하는 데 사용
targetYearweek = "2025/09"
orgDelimeter = "/"
newDelimeter = "_"
targetYearweek = targetYearweek.replace(orgDelimeter,newDelimeter)
print(targetYearweek)
```

```python
# 구분자를 기준으로 문자열 분리 : split()
targetYearweek = "2025W09"
targetYearweek = targetYearweek.upper()
specificChar = "W"

resultValue = targetYearweek.split(specificChar) # list(집합) 반환
type(resultValue)
```

```python
targetValue = 202522
# targetValue[0:4] -> 오류
# 인덱스 슬라이싱은 문자열에서만 가능 => 형변환 시켜야 함
targetValue = str(targetValue)
yearValue = targetValue[:4]
monValue = targetValue[4:]
print(yearValue,monValue)

# int형으로 변환해 나머지연산 쓸 수도 있음, 100자리 나머지 연산해서 남김
intValue = int(targetValue)
weekValue = intValue%100
print(weekValue)
print(intValue)
```
