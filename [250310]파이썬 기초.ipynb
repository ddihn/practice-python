```python
# 리스트 값 업데이트
cardList = ["우리V카드", "K-PASS 우리카드", "정석카드"]
cardList[0] = "NEW우리카드" # list update
print(cardList) # ['NEW우리카드', 'K-PASS 우리카드', '정석카드']

# 리스트 값 삽입
newCardInfo = "KB국민카드"
cardList.append(newCardInfo)
cardList.insert(3,"새로운카드") # 특정 인덱스에 값 삽입
print(cardList) # ['NEW우리카드', 'K-PASS 우리카드', '정석카드', 'KB국민카드']

# 리스트 값 삭제 ( 인덱스 활용, 값 활용 )
cardList.remove(cardList[0])
print(cardList) # ['K-PASS 우리카드', '정석카드', 'KB국민카드']

exceptList = ["정석카드"]
cardList.remove(exceptList[0])
print(cardList) # ['K-PASS 우리카드', 'KB국민카드']

# 리스트 정렬
cardList = ["가 카드", "나 카드","다 카드"]
scoreList = [100, 50, 70, 90]
scoreList.sort() #오름차순 정리, reverse = true면 내림차순
print(scoreList)
```

```python
# 리스트의 연산 ( 최솟값, 최댓값, 합계, 길이 ), 평균은 없음
min( scoreList )
max( scoreList )
sum( scoreList )
len( scoreList )
```
